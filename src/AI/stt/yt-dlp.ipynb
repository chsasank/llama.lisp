```python
import json
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import JSONFormatter
from tqdm.auto import tqdm
import time
import subprocess
```

```python
ytt_api = YouTubeTranscriptApi()
with open('kannada_asr_data.json') as kad:
    data = json.load(kad)
```

```python
#we shall try to get the transcripts for the url now 
total_failed = 0
failed_video_id = list()
for count, d_ict in enumerate(tqdm(data)):
    url = d_ict["video"]
    video_id  = url.split("=")[-1]
    try:
        trancript_of_video_id = ytt_api.fetch(video_id, languages=['kn'])
        formatter = JSONFormatter()
        json_formatted = formatter.format_transcript(trancript_of_video_id)
        json_formatted_dict = json.loads(json_formatted)
        data[count]['transcript'] = json_formatted_dict
        time.sleep(1)
    except:
        data[count]['transcript'] = None
        failed_video_id.append(video_id)
        total_failed = total_failed + 1    

with open('kannada_asr_data_with_transcript.json' , 'w+') as kadt:
    kadt.write(json.dumps(data))
    
print(f'\n total_failed is {total_failed}')
```
